```python
import random

#Classe básica de criaturas, bem aberto para futuras melhorias e herança
class Creature:

#Iniciador responsável por criar as criaturas com atributos obrigatórios
  def __init__(self, name, hp, save_mod):
    self.name = name
    self.hpmax = hp
    self.hp = hp
    self.save_mod = save_mod
    self.is_alive = True

# Método responsável por fazer o cálculo de dano e avaliar se a criatura morreu
  def takeDamage(self, amount):
    damage_left = amount - self.hp
    self.hp -= amount
    if self.hp <= 0:
      self.hp = 0
      self.is_alive = False
      if damage_left >= self.hpmax:
        print(f'The {self.name} takes {amount} of damage! His HP are now {self.hp}\nThe damage was too much, its a insta kill!')
      else:
        print(f'The {self.name} takes {amount} of damage! His HP are now {self.hp}\nIniciating the death saving throw')
  #O teste de morte se passa caso a criatura sofra danos o suficiente para ficar com 0 de HP
        death = []
        sucess = 0
        fail = 0
  #É iniciado uma melhor de 5, caso o d20 caia maior que 10 três vezes ele fica vivo, caso contrário morre
        for results in range(5):
          death_save = random.randint(1, 20)
          if death_save >= 11:
            sucess += 1
            if sucess == 3:
              print(f'The {self.name} resist his injuries, and is stabilized!')
              self.hp = 1
              self.is_alive = True
              break
          elif death_save <= 10:
            fail += 1
            if fail == 3:
              print(f'The {self.name} cannot resist his injuries, and is now dead!')
              break
    else:
      print(f'The {self.name} takes {amount} of damage! His HP are now {self.hp}')

#Classe de magias, igual criaturas, está propenso a futuras melhorias
class Spell:
  def __init__(self, name, damage_dice, dc):
    self.name = name #Nome da magia
    self.damage_dice = damage_dice #Tipo de dado e quantia EX.: 8d6
    self.dc = dc #Toda magia tem um numero para evitar o dano total

#Método responsável por rolar os dados de dano da magia
  def rollDamage(self):
#Este método separa o tipo de dado e a quantida a ser rolada, ao final teremos o dano total
    number, dice = map(int, self.damage_dice.lower().split('d'))
    return sum(random.randint(1,dice) for _ in range(number))

#Aqui juntamos as duas classes, e fazemos um calculo para ataques de magia em área
def cast_aoe_spell(spell, targets):
  print(f'\n-----Casting a {spell.name}, Spell save DC {spell.dc}-----')
  full_damage = spell.rollDamage()

  print(f'\nFull damage equals: {full_damage} ')
  for target in targets:
    if not target.is_alive:
      continue

    save_roll = random.randint(1, 20) + target.save_mod
    if save_roll >= spell.dc:
      final_damage = full_damage //2
      print(f'\nThe {target.name} pass the test! final damage: {final_damage}')

    else:
      final_damage = full_damage
      print(f'\nThe {target.name} fail the test! final damage: {final_damage}')

#Ele analisa alvo a alvo da lista, e se ele passou no teste toma somente metade do dano
    target.takeDamage(final_damage)

# --- TESTANDO O PROGRAMA ---
# Criando um grupo de Goblins
goblins = [
    Creature("Goblin 1", 6, 2),
    Creature("Goblin 2", 12, 2),
    Creature("Goblin Chief", 25, 3),
    Creature("Elf teammate", 35, 5),
    Creature("Barbarian teammate", 60, 3),
    Creature("Wizard teammate", 12, 2),
    Creature("Fighter teammate", 20, 3)
]

# Criando a famosa Fireball (Bola de Fogo)
fireball = Spell("Fireball", "8d6", 14)

# Executando a ação
cast_aoe_spell(fireball, goblins)

#Resumindo, estas linhas de código abrem diversas portas para melhorias futuras, adicionar mais caracteristicas de criaturas e resistencias elementais!
```
